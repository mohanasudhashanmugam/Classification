```python
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
dataset['classification'].value_counts()
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
independent=dataset.drop(['classification_yes'],axis=1)

independent
```

```python
dependent=dataset[['classification_yes']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
#from sklearn.preprocessing import StandardScaler
#sc=StandardScaler()
#x_train=sc.fit_transform(x_train)
#x_test=sc.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB

classifier_GNB=GaussianNB()

classifier_GNB.fit(x_train,y_train)
y_pred=classifier_GNB.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
print(cr)

from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,y_pred,average='weighted')

print('The f1_macro value', f1_macro)


from sklearn.metrics import roc_auc_score
roc_auc_score = roc_auc_score(y_test,classifier_GNB.predict_proba(x_test)[:,1])
print('roc_auc_score', roc_auc_score)
```

```python
from sklearn.naive_bayes import MultinomialNB

classifier_MNB=MultinomialNB()

classifier_MNB.fit(x_train,y_train)
y_pred=classifier_MNB.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
print(cr)

from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,y_pred,average='weighted')

print('The f1_macro value', f1_macro)


from sklearn.metrics import roc_auc_score
roc_auc_score = roc_auc_score(y_test,classifier_MNB.predict_proba(x_test)[:,1])
print('roc_auc_score', roc_auc_score)
```

```python
from sklearn.naive_bayes import BernoulliNB

classifier_BNB=BernoulliNB()

classifier_BNB.fit(x_train,y_train)
y_pred=classifier_BNB.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
print(cr)

from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,y_pred,average='weighted')
print('The f1_macro value', f1_macro)

from sklearn.metrics import roc_auc_score
roc_auc_score = roc_auc_score(y_test,classifier_BNB.predict_proba(x_test)[:,1])
print('roc_auc_score', roc_auc_score)
```
