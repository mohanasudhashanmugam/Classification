```python
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
dataset['classification'].value_counts()
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
independent=dataset.drop(['classification_yes'],axis=1)

independent
```

```python
dependent=dataset[['classification_yes']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

param_grid=[{'kernel': ['linear', 'poly', 'rbf', 'sigmoid']},
           {'gamma':['scale', 'auto']},{'C': [1, 10, 100]}]

grid=GridSearchCV(SVC(probability=True),param_grid,refit=True,cv=5,n_jobs=-1,verbose=3,scoring='f1_weighted')

#grid.fit(x_train,y_train)
```

```python
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
print(re)
```

```python
y_pred=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
```

```python
print(cr)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,y_pred,average='weighted')
print('The f1_macro value for best param:{}'.format(grid.best_params_), f1_macro)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
